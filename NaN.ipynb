Tratamento de dados faltantes

```python
import pandas as pd
dados = pd.read_csv('athlete_events.csv')
```

```python
dados.head()
```

```python
#exclui todas as linhas com NaN
dados2 = dados.dropna()
dados2.head()
```

```python
print(dados.shape)
print(dados2.shape)
```

```python
#para cada variável, qual valor é nulo ou não (boolean)
enulo = dados.isnull()
enulo.head()
```

```python
#Quantos valores faltantes para cada coluna
faltantes = dados.isnull().sum()
faltantes
```

```python
#coluna ID vai mostrar quantas linhas existem no dataframe
faltantes_percentual = (dados.isnull().sum() / len(dados['ID']))*100
faltantes_percentual
```

```python
#substituir dados faltantes
dados['Medal'].fillna('Nenhuma', inplace=True)
dados['Age'].fillna(dados['Age'].mean(), inplace = True)
dados['Height'].fillna(dados['Height'].mean(), inplace = True)
dados['Weight'].fillna(dados['Weight'].mean(), inplace = True)
dados.head()
```

```python
faltantes_percentual = (dados.isnull().sum() / len(dados['ID']))*100
print(faltantes_percentual)
```
