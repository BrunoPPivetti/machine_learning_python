Subplots

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(-1000,1000,1)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
y1 = x**2
y2 = x**3
y3 = -x**2
y4 = -x**3
#subplot divide em (linhas,colunas,lugar da divisão que o gráfico ficará)
#nesse exemplo, dividiu-se em 2 colunas e 2 linhas, ou seja, 4 posições diferentes
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.subplot(1,2,2)
plt.plot(x,y2)
plt.subplot(2,2,3)
plt.plot(x,y3)
#plt.subplot(2,2,4)
#plt.plot(x,y4) 
plt.show()
```

```python
y1 = x**2
y2 = x**3
y3 = -x**2
y4 = -x**3
y = [y1,y2,y3,y4]

figura = plt.figure(figsize=(10,10))
for i in range(4):
    figura.add_subplot(2,2,i+1) #soma 1 porque o python começa no índice 0
    plt.plot(x, y[i])
plt.show()
```
