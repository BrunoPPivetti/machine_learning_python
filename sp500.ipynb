Rodando um modelo de machine learning

```python
import pandas as pd
df = pd.read_csv('GSPC.csv')
df.head()
```

```python
df = df.drop('Date', axis=1)
```

```python
df[-2::]
```

```python
amanha = df[-1::]
amanha
```

```python
base = df.drop(df[-1::].index, axis=0)
base.tail()
```

```python
base['target'] = base['Close'][1:len(base)].reset_index(drop=True)
base.tail()
```

```python
prev = base[-1::].drop('target', axis=1)
prev
```

```python
treino = base.drop(base[-1::].index, axis=0)
treino.tail()
```

```python
#o intuito é prever se o dia terá um fechamento maior ou menor que o dia anterior ou seja, classificação
#para isso, a variável target precisa ser 1 ou 0, não uma variável escalar
#Caso 'target' seja maior que 'close' = 1
treino.loc[treino['target'] > treino['Close'], 'target'] = 1
treino.tail()
```

```python
treino.loc[treino['target'] != 1, 'target'] = 0
treino.tail()
```

```python
%pip install scikit-learn
```

```python
y = treino['target']
x = treino.drop('target', axis=1)

from sklearn.model_selection import train_test_split
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3)

from sklearn.ensemble import ExtraTreesClassifier
modelo = ExtraTreesClassifier()
modelo.fit(x_treino, y_treino)

resultado = modelo.score(x_teste, y_teste)
print("Acurácia:", resultado)
```

```python
print(resultado)
```

```python
modelo.predict(prev)
```

```python
prev['target'] = modelo.predict(prev)
prev
```
